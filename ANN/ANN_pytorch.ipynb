# Pytorch

### What is Pytorch?
- Open-Source deep learning framewok that provides flexibility and dynamic computation for building and training machine learning models
  
 - #### Core Components of PyTorch
   - **Tensors**: Multi-dimentional arrays similar to NumPy arrays but with GPU support fro accelerated computation
   - **Autograd**: Automatic differentiation engine that computes gradients for optimization
   - **torch.nn Module:** Provides tools to define and train neural networks with layers, activation function and loss function

#### Building a Neural Network in PyTorch
Steps
 - Define the Model
   - User `torch.nn.Module` to create a neutal network with layers and forward propagation
 - Define the Loss Function
   - Use build-in loss functions like Cross-Entropy Loss
 - Define the Optimizer
   - Use optmizers like SGD or Adam for weight updates

#### Training, Evaluaiting and Saving a Model in PyTorch
- **Training**
  - Forward pass to compute predictions
  - Compute loss and gradients using backpropagation
  - Update weights using an optimizer
- **Evaluation**
  - Test the model on Unseen data and calculate metrics like accuracy
- **Saving and Loading**
  - Save the model's parameter using `torch.save()` and load them using `torch.load()`

### Hands-On Exercise

Objective
 - Build, train, evaluate and save a neural netwoek for MNIST digit classification using PyTorch

```python
import torch

# torchvision - datasets: provide popular datasets like MNIST,CIPHER-10
# torchvision - trasformations for preprocessing data and normalising
from torchvision import datasets, transforms 

# Data-loader is a unitlity to handling (Data patching, shuffling and loading )
from torch.utils.data import DataLoader

# Provides classes and functions for building and defining neural network
# Contains predefined building blocks like layers, activation function and more
import torch.nn as nn

# Contains functions for operations that can be applied directly without defininf them as layers
# these are stateless functions, they dont haveparameter that needed to be learned (weights, biases).git/
# used for applying activaition and loss computation
import torch.nn.functional as F
```

### Transform.compose
Combines list of transformation to be applied sequentially
- `ToTensor:` Converting image data to pytorch tensor, which scales pixel walues to 0 to 1
- `Normalising`: Normalising tensor data to have a mean of 0.5 and standard deviation of 0.5

```python
# Define transformation
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,),(0.5,))
])
```

#### loading the dataset
- `root="./data"` loads the data in the root folder, in the folder name  **data**
- and same we are using for test

```python
# Load Dataset
train_dataset = datasets.MNIST(root="./data",train=True, transform=transform, download=True)
test_dataset = datasets.MNIST(root="./data", train= False, transform=transform, download = True)
```

`DataLoader:` it loads the data in batches for training and testing
- `batch_size = 32`: no of samples per batch
- `Shuffle=True`: Shuffles data to ensure randomness
  -  In test data we `shuffle = False`, to keep the data in order for evaluation

```python
# Create Data Loaders
train_dataset = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
# prining the size of the dataset
print(f"Training Data Size:{len(train_dataset)}")
print(f"Test Data Size: {len(test_dataset)}")
```

### Defining the Neural Network

```python
# Define the model
class NeuralNetwork(nn.module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(28*28,128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64,10)
        
    def forward(self, x):
        x = self.flatten(x)
        x =  F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
            
```
