## Introduction to Feature Enginnering

##### What is feature Enginering ?
  - Process of transforming raw data into meaningful inputs for machine learning models
  - It envolves creatingg modifying or selecting features that enhance model performance
##### Why is Feature Enginnering important ?
  - Improves model accuracy
  - Reduces Model Complexity
  - Enables Model Intrepretability
  - Handles Data Challenges

---


### Types of Feature


- Types of Features: Catogorical, Numerical, Ordinal

1. **Catogorical Features** : Represents Different categories or labels
   - Converts categorical data into numerical format
   - Encoding Techniques: One-Hot Encoding & Label Encoding 
  
2. **Numerical Encoding** :  Represents Continuous or discrete numbers
   - Ensures all features Contribute equally to the model
   - Preprecessing techniques: 
    - Scaling: Min/Max Scaling and Standardization
3. **Ordinal Features** : Categorical data with a meaningful order
   - Framous Encoding technique
     - Ordinal encoding

---

- **Scaling**
  - Ensures all features contibute equally to the model 
  - Techniques: Min-max Scaling | Standardization
- **Encoding**
  - Converts categorical data in to numerical data
  - Techniques: One-hot encoding| Label Encoding
- **Transformation**
  - Applies mathmatical function to modify features
  - Examples: Log Transformation| Polynomaial features
- **Feature Selection**
  - Reduces the number of input feature to improve model performance
  - Techniques: Statistical Methods| Recursicve Feature Elimination (REF)

### Selection featues based on their datatypes


```python
# Categorcial features
categorical_features = df.select_dtypes(include=["object"]).columns
print(categorical_features)
```

```python
# Numerical Features
numerical_features = df.select_dtypes(include=["int64"]).columns
float_features = df.select_dtypes(include=["float64"]).columns
print(numerical_features.to_list())
print(float_features.to_list())
```

---

```python
# Display Summary of Categorical features
print("categorical_features Summary")
for col in categorical_features:
    print(f"{col}:\n", df[col].value_counts(),"\n")
```
