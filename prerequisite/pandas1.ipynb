# Pandas For Data Manipulation
- It provides easy to use datastructure which is called series in data science

- **Series**: One dimentional labeled array capable of holding data of any type, 
- **Data Frame**: Two Dimentional labeled datastructure like a table

##### Loading the module

````python
import pandas as pd
````

## 1. Types of Data Structures

### Series

One dimentional labeled array, capable of holding data of any type

````python
# A series
s = pd.Series([10,20,30], index = ["a","b","c"])
print(s)
````

### Dataframe

Two dimentional data-structure like a table

````python
# Dictionary
data = {"Name": ["Alice","Bob"], "Age":[25,30]}

# Convert to dataframe
df = pd.DataFrame(data)

# printing the dataframe
print(df)
````

## 3. Viewing Data
  - You can also spcify the no of rows you want to see, by passin that no as a parameter example ` df.head(3)`

### **Head and Tail Function**

````python
# Top Few Rows  
print(df.head())

# Bottom few Rows
print(df.tail())
````

### **for Summary and Info About Data**

#### `df.info`
- Gives Summary of the dataframe

````python
print(df.info())
````

#### `df.describe();`
- Gives Statistical Summary of the dataframe

````python
print(df.describe());
````

### **Selecting and Indexing** by coloumn

````python
# Selecting Single Coloumn
print(df["Name"])

print("--------------------------")

# Selecting Multiple Conlumn
print(df[['Name','Age']])
````

### **Filtering Rows and Columns**
  we use `iloc ` which means index loacation

  - Way of writing for complex filtering

  - ```python
        rows = df[(df["sepal_width"]>4) & (df["species"] == "setosa")]
        print(rows)
    ```

````python
print(df[df["Age"]>25])
````

````python
# Will give first row by postion
print(df.iloc[0])
````

````python
# Will give first column by position
print(df.iloc[:,0])
````

````python
print(df.loc[0])
````

````python
print(df.loc[:0])
````
