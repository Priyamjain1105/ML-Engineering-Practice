# Data Aggregation and Grouping in Pandas

Grouping data allows yo to perform operations in subsets of data based on shared categories

### Creating the Group
```py
grouped = df.groupby("Column_name")
```


### Performing Different Operations on Group







##### 1. Iterate over group
```py
for name, group in grouped:
    print(name, group, sep="\n")
```
---


##### 2. Apply Aggregations
`grouped.mean()`
`grouped.sum()`

example
```py
df.groupby("category_column")["numeric_column"].mean()
```

````python
import numpy as np
import pandas as pd
# Student performance data
student_data = {
    'Class': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Gender': ['M', 'F', 'F', 'M', 'M', 'F', 'F', 'M'],
    'Score': [85, 90, 92, 78, 88, 95, 91, 82],
    'Subject': ['Math', 'Math', 'Science', 'Science', 
               'Math', 'Science', 'Math', 'Science']
}

df = pd.DataFrame(student_data)

# Average score by class
class_avg = df.groupby('Class')['Score'].mean()
print(class_avg)

# Top score by gender and subject
top_scores = df.groupby(['Gender', 'Subject'])['Score'].max()
print(top_scores)

# Number of students in each class-gender combination
student_counts = df.groupby(['Class', 'Gender']).size()
print(student_counts)
````

#### To apply many aggregation functiuons at column / Multi aggregation
```py
df.groupby("category_column".agg({"numeric_column":["mean","max", "min"],
  "numeric_column2: 'median
})
```


````python
# Sample data
data = {
    'Department': ['Sales', 'IT', 'Sales', 'HR', 'IT', 'HR'],
    'Salary': [50000, 60000, 55000, 45000, 65000, 48000],
    'Age': [32, 28, 41, 35, 27, 38]
}

df = pd.DataFrame(data)

# Multiple aggregations
result = df.groupby('Department').agg({
    'Salary': ['min', 'max', 'mean'],
    'Age': 'median'
})

print(result)
````

##### 4. Custom Aggregation Function

```py

    return x.max()-x.min()

df.groupby("Category_column")["numeric_column"].agg(range_func)
```

### Hands-on

````python
# Creating the dataset
import pandas as pd
data = {
    "Class": ["A","B","A","B","C","C"],
    "Score": [85, 90, 88, 73, 95, 80],
    "Age":   [15, 16, 15, 17, 16, 15]
}

df = pd.DataFrame(data)
print(df)
````

````python
# Mean by class
grouped = df.groupby("Class")["Score"].mean()
print(grouped)
````

````python
# Using Stats
stats = df.groupby("Class").agg({
    "Score":["mean","max","min"],
    "Age":["mean","max","min"]
    
})

print(stats)
````
