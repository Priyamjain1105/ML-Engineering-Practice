```python
import pandas as pd

# Sample data
data = {'Department': ['Sales', 'IT', 'Sales', 'HR', 'IT', 'HR'],
        'Employee': ['John', 'Jane', 'Mike', 'Sarah', 'Tom', 'Lisa'],
        'Salary': [50000, 60000, 55000, 45000, 65000, 48000]}

df = pd.DataFrame(data)

# Group by department and calculate average salary
avg_salary = df.groupby('Department')['Salary'].mean()
print(avg_salary)
```

```python
# Student performance data
student_data = {
    'Class': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Gender': ['M', 'F', 'F', 'M', 'M', 'F', 'F', 'M'],
    'Score': [85, 90, 92, 78, 88, 95, 91, 82],
    'Subject': ['Math', 'Math', 'Science', 'Science', 
               'Math', 'Science', 'Math', 'Science']
}

df = pd.DataFrame(student_data)

# Average score by class
class_avg = df.groupby('Class')['Score'].mean()
""" op: class mean-score
         a    ...
         b    ...
         
"""
print(class_avg)

# Top score by gender and subject
top_scores = df.groupby(['Gender', 'Subject'])['Score'].max()
"""
    Gender  Subject    Score(max)
      M        math       88
               Science  
      F        math
               Science

"""
print(top_scores)

# Number of students in each class-gender combination
student_counts = df.groupby(['Class', 'Gender']).size()
""" 
class gender
  A     M      4
        f      
"""
print(student_counts)
```

```python
import pandas as pd
import numpy as np

# Create sample dataset
np.random.seed(42)
data = {
    'order_id': range(1001, 1021),
    'customer_id': np.random.choice(['C100', 'C101', 'C102', 'C103', 'C104'], 20),
    'product_category': np.random.choice(['Electronics', 'Clothing', 'Home', 'Books'], 20),
    'order_date': pd.date_range('2023-01-01', periods=20, freq='D'),
    'order_amount': np.random.randint(50, 500, 20),
    'payment_method': np.random.choice(['Credit Card', 'PayPal', 'Bank Transfer'], 20),
    'customer_age': np.random.randint(18, 65, 20),
    'city': np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston'], 20)
}

df = pd.DataFrame(data)
print(df.head(5))
```

```python
totalbyproduct = df.groupby("product_category")["order_amount"].sum()
print(totalbyproduct)
```

```python
avgamt = df.groupby("payment_method")["order_amount"].mean()
print(avgamt)
```

```python
revenue = df.groupby(['city','product_category'])['order_amount'].sum()
print(revenue)
```

```python
cout_by_age = df.groupby('customer_age')['order_id'].size()
print(cout_by_age)
```

```python
import pandas as pd

# Sample sales data
data = {
    'Region': ['East', 'West', 'East', 'West', 'East', 'West'],
    'Product': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 125, 175, 225]
}

df = pd.DataFrame(data)

# Create pivot table showing average sales by region and product
pivot = pd.pivot_table(df, values='Sales', index='Region', columns='Product', aggfunc='mean')

print(pivot)
```

```python
# Get both sum and average sales
pivot = pd.pivot_table(df, values='Sales', index='Region', columns='Product', 
                      aggfunc=['sum', 'mean'])
print(pivot)
```

```python
# Add year to our sample data
df['Year'] = [2022, 2022, 2022, 2023, 2023, 2023]

# Pivot with multiple index and column levels
pivot = pd.pivot_table(df, values='Sales', index=['Region', 'Year'], 
                      columns='Product', aggfunc='sum',margins = True)

print(pivot)
```
