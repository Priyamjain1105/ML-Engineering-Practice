## Random Number Generation and Setting Seeds
### Random Number Generation
- np.random

```python
import numpy as np

#this will create the random array of 3x3 matrix, with uniform distribution from 0 to 1
random_array = np.random.rand(3,3)
print(random_array)
```

---

### Generating random integer
`np.random.randint(0,10, size=(2,3))`
will generate the matrix of 2x3
where integer ranging from 0 to 10

```python

random_integer = np.random.randint(0,10, size=(2,3))
print(random_integer)
```

Setting the Seed: You call np.random.seed(some_integer_value), where some_integer_value can be any integer, like 42. This initializes the random number generator with a specific state.

```python
np.random.seed(42)
random_integer = np.random.randint(0,10, size=(2,3))
print(random_integer)
```

### Question
- Generate the random dataset
- Replace the values which are greater then 25 to 0
- Calculate it stats like sum, mean, standard deviation

```python
#Generate Random Dataset
dataset = np.random.randint(1,51, size=(5,5))
print("Original Matrix",dataset, sep = "\n")
# Filter values > 25 and replace with 0
dataset[dataset>25] = 0;
print("Modified Dataset: \n", dataset)

#calculate summary stats
print("Sum:", np.sum(dataset))
print("Mean",np.mean(dataset))
print("Standard Deviation:", np.std(dataset))
```
