# Data Cleaning and Preparation with Pandas

## 1. Handeling Missing Values
Pandas provide tools to handel missing data

Methods to Handle Missing Values :
- Drop Missing Values
- Fill Missing Values
- Interpolation : fill data that is on similar lines to that particular data

### Droping Rows With Missing Values

##### To drop rows with missing values
```python
    df = df.dropna()
```

##### To drop Columns with missing values
```python
   df = df.dropna(axis=1)
```
---

### Filling Missing Values
- When 95% of the dataset is available and 5% is missing
- Here Filna (Fill Not Available's)
#### Simple filling the values in column 
```python
   df["column_name"] = df["column_name"].fillna(0)
```

### Forward Filling and Backward Filling
Filling w.r.t to forward values
   - `df.fillna(method="ffill")`
  
Filling w.r.t to backward values
   - `df.fillna(method="bfill")`

---

### Interpolate
`df["column_name"] = df["column_name"].interpolate()`


---

## 2. Data Transformations
- Renaming Columns
   - `df = df.rename(columns={"old_name":"new_name"})`
- Changing Data Types
   - `df["column_name"] = df["columns_name"].astyps("float")`
- Creating or Modifying Columns
   - `df["new_column"] = df["existing_column"]*2 `

---

## 3. Combining and Merging DataFrames
- Concatenation
  - To Combine Dataframe along Rows
    - `combined = pd.concat([df1, df2], axis=0)`
  - To Combine Dataframe aling Columns
    - `combined = pd.concat([df1, df2], axis=1)`
  ---
- Merging
  - Merging data based on the common_column
    - `merged = pd.merge(df1, df2, on="common_column")`
  - Left Join
    - `merged = pd.merge(df1, df2, how ="left", on="common_column" )`
  - Inner Join
    -  `merged = pd.merge(df1, df2, how="right", on = "common_coloumn")`
   ---
  
- Joining (just using index allignment without the column)
   - `joined df1.join(df2, how="inner")`

# Hands-On Exercise

### Cleaning the data by handeling the missing values and Renaming Columns

````python
#Importing the modules
import pandas as pd
import numpy as np
````

````python
# Creating the dataset
data = {
    "Name":["Alice","Bob", np.nan, "David"],
    "Age": [ 25, np.nan, 30, 35],
    "Score": [85, 90, np.nan, 88]
}

df = pd.DataFrame

df = pd.DataFrame(data)

print("Original Dataset \n", df)
````

````python
# Fillng the data using fillne
df["Age"]= df["Age"].fillna(df["Age"].mean())

#dInterpolating the Values
df["Score"] = df["Score"].interpolate()
````

````python
print("Dataset: \n", df)
````

````python
df.rename(columns={"Name":"Student_Name", "Score":"Exam_Score"})
````
