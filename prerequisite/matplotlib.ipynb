# Data Visualisation with Matplotlib and Seaborn 
Matplotlib: Static python librady to create static, interactive and animated plots, it allows for detailed cutomisation of plots

```python
import matplotlib.pyplot as plt

# Basic Plot
x = [1,2,3,4]
y = [10, 20, 25, 30]

plt.plot(x,y)
plt.show()
```

### Line Plot

```python
# line Plot
plt.plot([1,2,3], [10, 20, 30], label ="Trend", color ="orange", linestyle="--", marker="o")
plt.plot([1,2,3], [20, 30, 50], label ="Trend2")
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

---

<H4 style = "color:pink"> Bar-Chart</H4>
Used for Catagorical Data Comparison.

To visualise trends over time or sequences.


```python
categories = ["A","B","C"]
values = [10, 15, 7]

plt.bar(categories, values, color="blue")
plt.title("Bar Chart")

plt.show()
```

---

<H4 style = "color:yellow"> Histogram</H4>
Shows the distribution of the dataset



```python
data = [1,2,2,3,3,3,4,4,4,4]
plt.hist(data,bins=4,color="green",edgecolor="black")
plt.title("Histogram")
plt.show()
```

---

<H4 style = "color:skyblue"> Scatter Plot</H4>
It visualises the relationship btw two continuous variables

```python
x = [1,2,3,4,5]
y = [10, 12, 25, 40, 45]
plt.scatter(x,y, color="red")
plt.title("Scatter Plot")
plt.show()
```

---

## Customizing the plots

```python
x = [1,2,3,4,5]
y = [10, 12, 25, 40, 45]
plt.scatter(x,y, color="red")
plt.title("Scatter Plot")

# X-label
plt.xlabel("X-axis Label")

# Y-Label
plt.ylabel("Y-axis Label")

# Legend
plt.legend(["Dataset 1"])

plt.show()
```

### Hands-On

```python
# Creating the dataset
years = [2010, 2011, 2012, 2013]
sales = [100, 120, 140, 160]

# Bar chart
plt.plot(years, sales, label="Sales Trend", color="Blue", marker="o")
plt.title("Sales Over Years")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
#Bar chart
categories = ["Electronics","Clothing","Groceries"]
revenue = [250, 400, 150]

plt.bar(categories, revenue, color="blue")
plt.title("Revenue by Categories")
plt.show()
```

```python
# Scatter Plot
hours_studied = [1,2,3,4,5]
exam_scores = [50, 55, 65, 75, 85]

plt.scatter(hours_studied, exam_scores, color= "red")
plt.title("Relationship btw marks and study hours")
plt.xlabel("hours study")
plt.ylabel("Scored Marks")
plt.legend(["Student Scores"])
plt.show()
```
